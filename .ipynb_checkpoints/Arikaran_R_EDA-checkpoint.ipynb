```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
customers_df = pd.read_csv('X:/data_science_project/Data_Analysis/Customers.csv')
products_df = pd.read_csv('X:/data_science_project/Data_Analysis/Products.csv')
transactions_df = pd.read_csv('X:/data_science_project/Data_Analysis/Transactions.csv')


# Convert dates to datetime
customers_df['SignupDate'] = pd.to_datetime(customers_df['SignupDate'])
transactions_df['TransactionDate'] = pd.to_datetime(transactions_df['TransactionDate'])

# Extract year and month
customers_df['SignupYearMonth'] = customers_df['SignupDate'].dt.to_period('M')
transactions_df['TransactionYearMonth'] = transactions_df['TransactionDate'].dt.to_period('M')

# Merge data
merged_df = pd.merge(transactions_df, customers_df, on='CustomerID', how='left')
merged_df = pd.merge(merged_df, products_df, on='ProductID', how='left')

# 1. Retention Rate
merged_df['CustomerStatus'] = merged_df.apply(
    lambda row: 'New' if row['TransactionYearMonth'] == row['SignupYearMonth'] else 'Returning', axis=1
)
retention_rate = merged_df.groupby(['TransactionYearMonth', 'CustomerStatus']).size().unstack(fill_value=0)
retention_rate['RetentionRate'] = retention_rate['Returning'] / (retention_rate['New'] + retention_rate['Returning'])

print("\n1. Retention Rate:")
print(retention_rate)

# Plot Retention Rate
retention_rate['RetentionRate'].plot(kind='line', marker='o', figsize=(10, 6), color='blue')
plt.title('Retention Rate Over Time')
plt.xlabel('Transaction Year-Month')
plt.ylabel('Retention Rate')
plt.grid(True)
plt.show()

# 2. Revenue by Product Category
revenue_by_category = merged_df.groupby('Category')['TotalValue'].sum()

print("\n2. Revenue by Product Category:")
print(revenue_by_category)

# Plot Revenue by Product Category
revenue_by_category.plot(kind='bar', figsize=(8, 6), color='green')
plt.title('Revenue by Product Category')
plt.xlabel('Category')
plt.ylabel('Total Revenue (USD)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# 3. Seasonal Trends by Category
seasonal_trends = merged_df.groupby(['TransactionYearMonth', 'Category'])['TotalValue'].sum().unstack(fill_value=0)

print("\n3. Seasonal Trends by Category:")
print(seasonal_trends)

# Plot Seasonal Trends
seasonal_trends.plot(kind='line', figsize=(12, 8))
plt.title('Seasonal Trends by Category')
plt.xlabel('Transaction Year-Month')
plt.ylabel('Revenue (USD)')
plt.legend(title='Category')
plt.grid(True)
plt.show()

# 4. High-Value Customers (Top 10% contributors)
total_revenue_by_customer = merged_df.groupby('CustomerID')['TotalValue'].sum()
high_value_threshold = total_revenue_by_customer.quantile(0.90)
high_value_customers = total_revenue_by_customer[total_revenue_by_customer >= high_value_threshold]

print("\n4. High-Value Customers (Top 10%):")
print(high_value_customers)

# Plot High-Value Customers
high_value_customers.sort_values(ascending=False).plot(kind='bar', figsize=(10, 6), color='orange')
plt.title('High-Value Customers (Top 10%)')
plt.xlabel('Customer ID')
plt.ylabel('Total Revenue (USD)')
plt.grid(axis='y')
plt.show()

# 5. Revenue Concentration (Top 10% Transactions)
print("\n5. Revenue Concentration (Top 10% Transactions")
top_10_percent_revenue = merged_df[merged_df['TotalValue'] >= top_10_percent_transactions]['TotalValue'].sum()
remaining_revenue = merged_df['TotalValue'].sum() - top_10_percent_revenue

# Plotting Revenue Concentration as a Pie Chart
plt.figure(figsize=(6, 6))
plt.pie(
    [top_10_percent_revenue, remaining_revenue],
    labels=['Top 10% Transactions', 'Remaining Transactions'],
    autopct='%1.1f%%',
    startangle=90,
    colors=['#66b3ff', '#99ff99']
)
plt.title('Revenue Concentration (Top 10% vs. Remaining)')
plt.show()

# 6. Signup Conversion Rate
print("\n6. Signup Conversion Rate")
signup_count = customers_df['CustomerID'].nunique()
converted_count = merged_df['CustomerID'].nunique()

# Bar Chart for Signup Conversion
plt.figure(figsize=(6, 6))
plt.bar(['Signups', 'Conversions'], [signup_count, converted_count], color=['#ff9999', '#66b3ff'])
plt.title('Signup Conversion Rate')
plt.ylabel('Number of Customers')
plt.show()

# 7. Top Regions by Category
revenue_by_region_category = merged_df.groupby(['Region', 'Category'])['TotalValue'].sum().unstack(fill_value=0)

print("\n7. Top Regions by Category:")
print(revenue_by_region_category)

# Plot Revenue by Region and Category
revenue_by_region_category.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='viridis')
plt.title('Revenue by Region and Product Category')
plt.xlabel('Region')
plt.ylabel('Total Revenue (USD)')
plt.xticks(rotation=45)
plt.legend(title='Category')
plt.grid(axis='y')
plt.show()

# 8. Monthly Average Order Value (AOV)
monthly_aov = merged_df.groupby('TransactionYearMonth')['TotalValue'].mean()

print("\n8. Monthly Average Order Value (AOV):")
print(monthly_aov)

# Plot Monthly AOV
monthly_aov.plot(kind='line', marker='o', figsize=(10, 6), color='purple')
plt.title('Monthly Average Order Value (AOV)')
plt.xlabel('Transaction Year-Month')
plt.ylabel('Average Order Value (USD)')
plt.grid(True)
plt.show()

# 9. Customer Frequency Segments
purchase_frequency = merged_df.groupby('CustomerID').size()
customer_frequency_segments = pd.cut(purchase_frequency, bins=[0, 1, 5, 10, float('inf')], labels=['1', '2-5', '6-10', '10+']).value_counts()

print("\n9. Customer Frequency Segments:")
print(customer_frequency_segments)

# Plot Customer Frequency Segments
customer_frequency_segments.plot(kind='bar', figsize=(8, 6), color='red')
plt.title('Customer Frequency Segments')
plt.xlabel('Segment')
plt.ylabel('Number of Customers')
plt.grid(axis='y')
plt.show()

# 10. Signup-Revenue Correlation
print("\n10. Signup-Revenue Correlation")
monthly_signups = customers_df.groupby('SignupYearMonth').size()
monthly_revenue = merged_df.groupby('SignupYearMonth')['TotalValue'].sum()

# Scatter Plot for Signup-Revenue Correlation
plt.figure(figsize=(8, 6))
plt.scatter(monthly_signups, monthly_revenue, color='purple')
plt.title('Signup-Revenue Correlation')
plt.xlabel('Number of Signups')
plt.ylabel('Total Revenue (USD)')
plt.grid(True)
plt.show()
```
